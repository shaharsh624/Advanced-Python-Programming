```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
distance_car = np.array([20, 30, 40, 60, 80, 90, 100])
distance_bike = np.array([30, 42, 55, 70, 72, 85, 100])
time = np.array([30, 42, 55, 70, 72, 86, 95])
```

```python
# Line - Distance vs Time

plt.xlabel("Distance")
plt.ylabel("Time")
plt.title("Distance vs Time Graph")

plt.plot(distance_car, time, label="Car")
plt.plot(distance_bike, time, label="Bike")

plt.legend()
```

```python
data = pd.read_csv("cwc2023standings.csv")
data
```

```python
# Bar - Team vs Points

plt.figure(figsize=(12,7))
plt.bar(data['TEAM'], data['POINTS'], color='skyblue')
plt.title("ICC CWC 2023 Points Table")
plt.xlabel("Team", fontsize=12)
plt.ylabel("Points", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Bar - Team vs NET RR

plt.figure(figsize=(12,7))
plt.bar(data['TEAM'], data['NET RR'], color='lightcoral')
plt.title("ICC CWC 2023 Points Table")
plt.xlabel("Team", fontsize=12)
plt.ylabel("Net Run Rate", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Pie - Points

plt.figure(figsize=(12,7))

colors = plt.cm.Paired(range(10))
plt.pie(data['POINTS'], labels=data['TEAM'], colors=colors, autopct='%1.1f%%')

plt.title("ICC CWC 2023 Points Table")

plt.show()
```

```python
# Pie - Points

plt.figure(figsize=(12,7))

colors = plt.cm.Paired(range(10))
plt.pie(data['POINTS'], labels=data['TEAM'], colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))

plt.title("ICC CWC 2023 Points Table")

plt.show()
```
