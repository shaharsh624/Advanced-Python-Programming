```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load your second dataset (replace your dataset path with your actual dataset file path)
df = pd.read_csv("\two_dataset.csv")
```

```python
sns.set(style="whitegrid")
```

```python
# 1. Unique Categorical Values 
unique_categories = df['category'].unique()
print("Unique Category Values: ", unique_categories)
unique_subcategories = df['subcategory'].unique()
print("Unique Sub-Category Values: ", unique_subcategories)
```

```python
# 2. Null Count
null_count = df.isnull().sum()
print("\nNull Count:")
print(null_count)
```

```python
# 3. Fill Null Values 
df['target_variable'].fillna(0, inplace=True)
```

```python
# 4. Pair Plot 
numerical_columns = df.select_dtypes(include=['float64', 'int64'])
sns.pairplot(numerical_columns)
plt.show()
```

```python
# 5. Histogram 
sns.histplot(df['value'], bins=20, kde=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Value')
plt.show()
```

```python
# 6. Boxplot 
sns.boxplot(x='category', y='value', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Correlation Matrix and Heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# 8. Bar Plot 
sns.barplot(x='category', y='value', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Line Plot 
sns.lineplot(x='date', y='value', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Scatter Plot
sns.scatterplot(x='date', y='value', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# 11. Violin Plot
sns.violinplot(x='category', y='value', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# 12. Count Plot
sns.countplot(x='category', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# 13. Missing Value Heatmap
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
```
