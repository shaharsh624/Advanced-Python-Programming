```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
distance_car = np.array([20, 30, 40, 60, 80, 90, 100])
distance_bike = np.array([30, 42, 55, 70, 73, 85, 100])
time = np.array([30, 42, 55, 70, 73, 86, 95])
```

```python
plt.xlabel("Distance in KMs")
plt.ylabel("Time in minutes")
plt.title("Distance vs Time Graph")

plt.plot(distance_car, time, label="Car")
plt.plot(distance_bike, time, label="Bike")

plt.legend()
plt.show()
```

```python
data = pd.read_csv("cwc2023standings.csv")
```

```python
data.head()
```

```python
data
```

```python
plt.figure(figsize=(12,7))
plt.bar(data['TEAM'], data['POINTS'], color='skyblue')

plt.title("ICC CWC 2023 Points Table")
plt.xlabel("Team", fontsize=14)
plt.ylabel("Points", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.bar(data['TEAM'], data['NET RR'], color='lightcoral')

plt.title("ICC CWC 2023 NRR Table")
plt.xlabel("Team", fontsize=14)
plt.ylabel("NRR", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))

colors = plt.cm.Paired(range(10))
plt.pie(data['POINTS'], labels=data['TEAM'], colors=colors)

plt.title("ICC CWC 2023 Points Table")

plt.show()
```

```python
plt.figure(figsize=(12,7))

colors = plt.cm.Paired(range(10))
plt.pie(data['POINTS'], labels=data['TEAM'], colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))

plt.title("ICC CWC 2023 Points Table")

plt.show()
```

```python
import seaborn as sns

# Set the Seaborn theme and context for better aesthetics
sns.set_theme(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

# Set the figure size
plt.figure(figsize=(14, 10))

sorted_data = data.sort_values(by=['POINTS', 'NET RR'], ascending=[False, False])

# Create horizontal bar plot for Points using Seaborn
sns.barplot(x=sorted_data['POINTS'], y=sorted_data['TEAM'], color='cyan', label='Points', orient='h')

# Overlay the NET RR on the same plot
sns.barplot(x=sorted_data['NET RR'], y=sorted_data['TEAM'], color='magenta', label='NET RR', orient='h')

# Add legend and title
plt.legend(loc='upper right')
plt.title('Points and Net Run Rate (NET RR) of Teams in CWC2023', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Team', fontsize=14)

# Display the plot
plt.show()
```
